```python


# load the house images and then scale the pixel intensities to the
# range [0, 1]
print("[INFO] loading urban images...")
images = datasets.load_urban_images(df, inputPathIMG )
images = images / 255.0

#print(images.shape)









```

```python
# construct a training and testing split with 75% of the data used
# for training and the remaining 25% for evaluation
print("[INFO] constructing training/testing split...")
#(train, test) = train_test_split(df, test_size=0.25, random_state=42)

# partition the data into training and testing splits using 75% of
# the data for training and the remaining 25% for testing
split = train_test_split(df3, images, test_size=0.25, random_state=42)
(trainAttrX, testAttrX, trainImagesX, testImagesX) = split
```

```python
trainY_mat = trainAttrX["MaterialCO2"]
testY_mat = testAttrX["MaterialCO2"]

trainY_op = trainAttrX["OperationCO2"]
testY_op = testAttrX["OperationCO2"]

trainY_c3c4 = trainAttrX["C3C4 MaterialCO2"]
testY_c3c4 = testAttrX["C3C4 MaterialCO2"]

trainY_d = trainAttrX["D Material CO2"]
testY_d = testAttrX["D Material CO2"]
```

```python
# process the house attributes data by performing min-max scaling
# on continuous features, one-hot encoding on categorical features,
# and then finally concatenating them together
print("[INFO] processing data...")
(trainAttrX, testAttrX) = datasets.process_building_attributes(df,trainAttrX, testAttrX)



# create our MLP and then compile the model using mean absolute
# percentage error as our loss, implying that we seek to minimize
# the absolute percentage difference between our price *predictions*
# and the *actual prices*
#model = models.create_mlp(trainX.shape[1], regress=True)
#opt = Adam(lr=1e-3, decay=1e-3 / 200)
#model.compile(loss="mean_absolute_percentage_error", optimizer=opt)

# train the model
#print("[INFO] training model...")
#model.fit(trainX, trainY, validation_data=(testX, testY),
#	epochs=200, batch_size=8)


print("[INFO] creating model...")
# create the MLP and CNN models
mlp = models.create_mlp(trainAttrX.shape[1], regress=False)
cnn = models.create_cnn(64, 64, 3, regress=False)
```

```python


#create the input to our final set of layers as the *output* of both
# the MLP and CNN
combinedInput = concatenate([cnn.output, mlp.output])

# our final FC layer head will have two dense layers, the final one
# being our regression head
x_op = Dense(64, activation="relu")(combinedInput)
x_op = Dense(32, activation="relu")(x_op)
x_op = Dense(16, activation="relu")(x_op)
x_op = Dense(8, activation="relu")(x_op)
x_op = Dense(1, activation="linear", name="operation_output")(x_op)

x_mat = Dense(64, activation="relu")(combinedInput)
x_mat = Dense(32, activation="relu")(x_mat)
x_mat = Dense(16, activation="relu")(x_mat)
x_mat = Dense(8, activation="relu")(x_mat)
x_mat = Dense(1, activation="linear", name="material_output")(x_mat)

x_c3c4 = Dense(64, activation="relu")(combinedInput)
x_c3c4 = Dense(32, activation="relu")(x_c3c4)
x_c3c4 = Dense(16, activation="relu")(x_c3c4)
x_c3c4 = Dense(8, activation="relu")(x_c3c4)
x_c3c4 = Dense(1, activation="linear", name="c3c4_output")(x_c3c4)

x_d = Dense(64, activation="relu")(combinedInput)
x_d = Dense(32, activation="relu")(x_d)
x_d = Dense(16, activation="relu")(x_d)
x_d = Dense(8, activation="relu")(x_d)
x_d = Dense(1, activation="linear", name="d_output")(x_d)

# our final model will accept categorical/numerical data on the MLP
# input and images on the CNN input, outputting a single value (the
# predicted price of the house)
model = Model(inputs=[cnn.input, mlp.input], outputs=[x_op, x_mat, x_c3c4, x_d])

```

```python


losses = {
	"operation_output": "mean_absolute_percentage_error",
	"material_output": "mean_absolute_percentage_error",
	"c3c4_output": "mean_absolute_percentage_error",
	"d_output": "mean_absolute_percentage_error",
}

# compile the model using mean absolute percentage error as our loss,
# implying that we seek to minimize the absolute percentage difference
# between our price *predictions* and the *actual prices*
opt = Adam(lr=1e-3, decay=1e-3 / 400)
model.compile(loss=losses, loss_weights=[1.0, 0.2, 0.2, 0.2], optimizer=opt)



# train the model
print("[INFO] training  CO2 model...")
history = model.fit(
	[trainImagesX, trainAttrX], [trainY_op, trainY_mat, trainY_c3c4, trainY_d],
	validation_data=([testImagesX,testAttrX], [testY_op, testY_mat, testY_c3c4, testY_d]),
	epochs=300, batch_size=25)
# make predictions on the testing data
#print("[INFO] predicting CO2...")
#preds = model.predict([testAttrX, testImagesX])
#print(preds)





```

```python

# compute the difference between the *predicted* house prices and the
# *actual* house prices, then compute the percentage difference and
# the absolute percentage difference
diff_op = preds[0].flatten() - testY_op
percentDiff_op = (diff_op / testY_op) * 100
absPercentDiff_op = np.abs(percentDiff_op)

# compute the mean and standard deviation of the absolute percentage
# difference
mean_op = np.mean(absPercentDiff_op)
std_op= np.std(absPercentDiff_op)


diff_mat = preds[1].flatten() - testY_mat
percentDiff_mat = (diff_mat / testY_mat) * 100
absPercentDiff_mat = np.abs(percentDiff_mat)

# compute the mean and standard deviation of the absolute percentage
# difference
mean_mat = np.mean(absPercentDiff_mat)
std_mat= np.std(absPercentDiff_mat)


diff_c3c4  = preds[2].flatten() - testY_c3c4
percentDiff_c3c4 = (diff_c3c4 / testY_c3c4) * 100
absPercentDiff_c3c4 = np.abs(percentDiff_c3c4)

# compute the mean and standard deviation of the absolute percentage
# difference
mean_c3c4 = np.mean(absPercentDiff_c3c4)
std_c3c4= np.std(absPercentDiff_c3c4)



diff_d = preds[3].flatten() - testY_d 
percentDiff_d  = (diff_d  / testY_d ) * 100
absPercentDiff_d = np.abs(percentDiff_d )

# compute the mean and standard deviation of the absolute percentage
# difference
mean_d  = np.mean(absPercentDiff_d )
std_d = np.std(absPercentDiff_d )



# finally, show some statistics on our model
#locale.setlocale(locale.LC_ALL, "en_US.UTF-8")

print("[INFO] MAT mean: {:.2f}%, std: {:.2f}%".format(mean_mat, std_mat))

print("[INFO] OP mean: {:.2f}%, std: {:.2f}%".format(mean_op, std_op))

print("[INFO] C3C4 mean: {:.2f}%, std: {:.2f}%".format(mean_c3c4, std_c3c4))

print("[INFO] D mean: {:.2f}%, std: {:.2f}%".format(mean_d, std_d))

#model.save(os.path.sep.join([args["dataset"], "model-opreg.h5"]))



```
